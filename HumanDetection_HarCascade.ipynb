```
import cv2
import time
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


#not good accuracy with group of people and less resolution images

person_cascade = cv2.CascadeClassifier(os.path.join('full_body.xml'))

frame = cv2.imread('download.jpg')
       
start_time = time.time()
frame = cv2.resize(frame,(640,360)) # Downscale to improve frame rate
gray_frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY) # Haar-cascade classifier needs a grayscale image
rects = person_cascade.detectMultiScale(gray_frame)
        
end_time = time.time()
print("Elapsed Time:",end_time-start_time)
  
  
if(len(rects) == 0):print("absence of human")

else:print("human present",len(rects))

  
  
for (x, y, w, h) in rects:
  cv2.rectangle(frame, (x,y), (x+w,y+h),(0,255,0),2)
  

plt.imshow(frame)
```
